```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import os

# Display the envelope array
# print(envelope)
def get_envelope(n, sample_rate):
    # Generate n random numbers from 0 to 1
    rand_numbers = np.random.rand(n+1)

    # Initialize an empty list to hold linspace arrays
    envelope_parts = []

    # Generate concatenated linspace arrays
    for i in range(n):
        # Each linspace goes from rand_number[i] to rand_number[i]+1 with int(1000/n) points
        part = np.linspace(rand_numbers[i], rand_numbers[i+1], int(sample_rate/n), endpoint=False)
        envelope_parts.append(part)

    # Concatenate all parts
    envelope = np.concatenate(envelope_parts)

    # Adjust the length of the envelope to exactly 1000
    if len(envelope) > sample_rate:
        envelope = envelope[:sample_rate]
    elif len(envelope) < sample_rate:
        # If less than 1000, extend by repeating the last value
        last_val = envelope[-1]
        envelope = np.concatenate([envelope, last_val * np.ones(1000 - len(envelope))])
        
    return envelope

def generate_oscillation(frequency, duration, sample_rate, envelope):
    """
    Generate a neuronal oscillation wave with a fluctuating envelope.

    Parameters:
        frequency (float): The frequency of the oscillation in Hz.
        duration (float): Duration of the signal in seconds.
        sample_rate (int): The sampling rate in Hz.
        amplitude (float): The maximum amplitude of the oscillation.

    Returns:
        np.ndarray, np.ndarray: Time array and the generated signal.
    """
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    # Basic sinusoidal signal
    wave = np.sin(2 * np.pi * frequency * t)
    # Create an envelope with 1Hz frequency to modulate the amplitude
    # envelope = 1 + 0.5 * np.sin(2 * np.pi * 1 * t)
    # Modulate the amplitude of the wave
    modulated_wave = wave * envelope
    return t, modulated_wave

colors = ['purple', 'blue', 'green', 'orange', 'red']
sample_rate = 10000  # 1000 samples per second
```

```python
# Parameters
frequency = 2  # 2 Hz delta wave
duration = 1  # 1 second

# envelope1 = np.linspace(1, 0.5, 100, endpoint=False)
# envelope2 = np.linspace(0.5, 0.7, 400, endpoint=False)
# envelope3 = np.linspace(0.7, 0.3, 200, endpoint=False)
# envelope4 = np.linspace(0.3, 0.9, 200, endpoint=False)
# envelope5 = np.linspace(0.9, 0.7, 100, endpoint=False)
# envelope = np.concatenate([envelope1, envelope2, envelope3, envelope4, envelope5])
# Generate the wave
envelope = get_envelope(4,sample_rate)
t, signal = generate_oscillation(frequency, duration, sample_rate, envelope)

# Plotting
cm = 1/2.54  # centimeters in inches
plt.figure(figsize=(13*cm, 4*cm))
plt.plot(t, signal, colors[0], linewidth=1.5)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (a.u.)')
plt.title('Delta oscillation')
# plt.legend()
plt.grid(False)
matplotlib.rcParams.update({'font.size': 10})
save_path = os.path.join('..', '..', '..', '..', 'DThesis', 'chapter1', 'delta.eps')
plt.savefig(save_path, format='eps')
plt.show()
```
